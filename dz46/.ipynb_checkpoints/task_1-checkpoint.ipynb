```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from adtk.visualization import plot
from adtk.detector import OutlierDetector

from sklearn.neighbors import LocalOutlierFactor
from sklearn.metrics import mean_absolute_error, mean_squared_error

from pmdarima import auto_arima

from statsmodels.graphics.tsaplots import plot_pacf,plot_acf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA

import warnings
warnings.filterwarnings("ignore")

# Функція для побудови графіків сезонності

def sesonal(data, s):
    plt.figure(figsize=(20,20), dpi= 80)
    for i, y in enumerate(data.index.year.unique()):
        plt.plot(list(range(1,len(data[data.index.year==y])+1)), data[data.index.year==y][data.columns[0]].values, label=y)
    plt.title("Сезонність по періодам")
    plt.legend(loc="right")
    plt.show()
```

```python
data = pd.read_csv("cardox.csv", index_col = "date", parse_dates = True)
data
```

```python
plot(data, ts_linewidth=2)
```

###### Тренд та сезонність присутні

```python
data.describe()
```

###### Досліджуємо кореляцію

```python
plot_acf(data, lags=50)
plt.title('Автокореляція')
plt.show()

plot_pacf(data, lags=50, method='ywm')
plt.title('Часткова автокореляція')
plt.show()
```

##### Досліджуємо сезонність

```python
sesonal(data,12)
```

```python
##### Виявляємо аномалії та виводимо їх на графік та у таблиці
```

```python
outlier_detector = OutlierDetector(LocalOutlierFactor())
anomalies = outlier_detector.fit_detect(data)
plot(data, anomaly=anomalies, ts_linewidth=1, anomaly_color='magenta', anomaly_alpha=0.3, curve_group='all');
anomalies[anomalies]
```

#### Коригуємо аномалії заміною аномалій на дані аналогічних місяців за сусідній рік без аномалій

```python
data.loc['1958-09-01']=data.loc['1960-09-01']
data.loc['1958-10-01']=data.loc['1960-10-01']
data.loc['1958-11-01']=data.loc['1960-11-01']
data.loc['1959-10-01']=data.loc['1960-10-01']
```

#### Проведення декомпозиції, коригування ряду і заповнення пропущених значень

```python
result_M = seasonal_decompose(data.value, model='multiplicative', extrapolate_trend=12)
plt.rcParams.update({'figure.figsize': (10,10)})
result_M.plot().suptitle('');

# перевірка на наявність пустих значень
if data['value'].isna().sum() == 0:
    print("Пропущені значення відсутні")
else:
    print("Необхідно провести нормалізацію")
```

#### Знаходження оптимальної моделі

```python
model = auto_arima(data.value, seasonal=True,
                   stepwise=True, trace=True, m=12)
print(model.summary())
```

```python
# визначено модель ARIMA(2,1,1)(1,0,1)[12]  
# тренуємо модель і робимо прогноз

train = data.iloc[:-12]
test = data.iloc[-12:]
model = ARIMA(train['value'], order=(2,1,1), seasonal_order=(1,0,1,12))
model_fit = model.fit()
forecast = model_fit.forecast(len(test), alpha=0.05)

plt.figure(figsize=(12,5))
plt.plot(train.index,train['value'], label='Train')
plt.plot(test.index,test['value'], label='Test')
plt.plot(test.index,forecast, label='Forecast')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Volume')
plt.legend()
plt.show()
```

#### Відображаємо діагностику моделі

```python
mae = mean_absolute_error(test['value'], forecast)
rmse = np.sqrt(mean_squared_error(test['value'], forecast))

print('MAE:', mae)
print('MSE:', rmse)

model_fit.plot_diagnostics(figsize=(15, 10))
plt.show()
```

#### Будуємо прогноз


```python
pred_uc = model_fit.get_forecast(steps=30*12)
pred_ci = pred_uc.conf_int()

ax = data['value'].plot(label='Actual', figsize=(14, 7))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Year')
ax.set_ylabel('Sales')
ax.set_title('Volume')
plt.legend()
plt.show()
```

```python
forecast = pd.DataFrame({'Predict': pred_uc.predicted_mean,
                           'Low': pred_ci['lower value'],
                           'Up': pred_ci['upper value']})
forecast.to_csv('forecast.csv')
```
