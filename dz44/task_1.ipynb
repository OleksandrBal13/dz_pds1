```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
import numpy as np
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
sns.displot(df, x='Age', kde=True)
plt.axvline(x=df.Age.median(), color='blue', label='Median')
plt.axvline(x=df.Age.mean(), color='red', label='Mean')
plt.legend(loc = 'upper right')
plt.title('Діаграма розподілу змінної Age')
```

```python
df.info()
df.isnull().sum()
```

```python
# Нормалізуємо параметр "Cabin"
df['Cabin'] = np.where(df['Cabin'].isnull(), 0, 1)
```

```python
# Перший метод заповнення - середнє значення
df['Age_Mean'] = df.copy()[['Age']]
df['Age_Mean'].fillna(df['Age'].mean(), inplace = True)

# Другий метод - медіанне значення
df['Age_Median'] = df.groupby(['Pclass', 'Sex'], group_keys=False)['Age'].apply(lambda x: x.fillna(x.median()))

# Третій метод - bfill + ffill
df['Age_B+F'] = df.groupby(['Pclass', 'Sex'], group_keys=False)['Age'].apply(lambda x: x.fillna(x.bfill().ffill()))
df
```

```python
# Графічно представимо три методи заповнення пропусків.

fig, axs = plt.subplots(1, 3)
sns.histplot(data=df, x="Age_Mean", color="Grey", kde=True, ax=axs[0])
sns.histplot(data=df, x="Age_Median", color="Black", kde=True, ax=axs[1])
sns.histplot(data=df, x="Age_B+F", color="Brown", kde=True, ax=axs[2])
plt.show()
```

```python
# перевірка гіпотез
from scipy.stats import shapiro
data = df['Age_Median']
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Гіпотезу Age_Median прийнято')
else:
    print('Гіпотезу Age_Median не прийнято')

data = df['Age_Mean']
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Гіпотезу Age_Mean прийнято')
else:
    print('Гіпотезу Age_Mean не прийнято')
    
data = df['Age_B+F']
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Гіпотезу Age_B+F прийнято')
else:
    print('Гіпотезу Age_B+F не прийнято')
```

Жоден з методів не пройшов перевірку на нормальність розподілення, але найкращий результат показав метод bfill + ffill
