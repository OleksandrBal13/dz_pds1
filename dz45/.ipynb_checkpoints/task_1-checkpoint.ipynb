```python
import pandas as pd
import numpy as np
import itertools
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from pmdarima.arima import auto_arima
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

```python
sunspot = sm.datasets.sunspots.load_pandas().data
sunspot
```

```python
plt.figure(figsize=(9,4))
plt.plot(sunspot['YEAR'],sunspot['SUNACTIVITY'])
plt.xlabel('Year')
plt.ylabel('Sunspot episodes per one year')
plt.show()
```

```python
plot_acf(sunspot['SUNACTIVITY'], lags=50)
plt.show()
plot_pacf(sunspot['SUNACTIVITY'], lags=50, method='ywm')
plt.show()
```

```python
model = auto_arima(sunspot['SUNACTIVITY'], seasonal=False, stepwise=True, trace=True)
print(model.summary())
```

```python
model.plot_diagnostics(figsize=(15, 12))
plt.show()
```

```python
train = sunspot.iloc[:-12]
test = sunspot.iloc[-12:]
#Ð·Ð°Ð·Ð½Ð°Ñ‡Ð°Ñ”Ð¼Ð¾ Ð¿Ð¾Ñ€ÑÐ´Ð¾Ðº Ð²Ñ–Ð´Ð¿Ð¾Ð²Ñ–Ð´Ð½Ð¾ Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚Ñƒ Ð²Ð¸Ð±Ð¾Ñ€Ñƒ Ð¼Ð¾Ð´ÐµÐ»Ñ–
model = ARIMA(train['SUNACTIVITY'], order=(2,1,3))
model_fit = model.fit()

# ÐŸÑ€Ð¾Ð³Ð½Ð¾Ð·
forecast = model_fit.forecast(len(test), alpha=0.05)

# Ð’Ñ–Ð´Ð¾Ð±Ñ€Ð°Ð¶ÐµÐ½Ð½Ñ Ð½Ð° Ð³Ñ€Ð°Ñ„Ñ–ÐºÑƒ
plt.figure(figsize=(9,4))
plt.plot(train['YEAR'],train['SUNACTIVITY'], label='Train')
plt.plot(test['YEAR'],test['SUNACTIVITY'], label='Test')
plt.plot(test['YEAR'],forecast, label='Forecast')
plt.title('Sunspot Forecast')
plt.xlabel('Year')
plt.ylabel('Sunspot episodes per one year')
plt.legend()
plt.show()
```

```python
# ÐŸÑ€Ð¾Ð³Ð½Ð¾Ð· Ð´Ð¾ 2030 Ñ€Ð¾ÐºÑƒ
model = ARIMA(sunspot['SUNACTIVITY'], order=(2,1,3))
model_fit = model.fit()
forecast = model_fit.forecast(21, alpha=0.05)

# Ð’Ñ–Ð´Ð¾Ð±Ñ€Ð°Ð¶ÐµÐ½Ð½Ñ Ð½Ð° Ð³Ñ€Ð°Ñ„Ñ–ÐºÑƒ
plt.figure(figsize=(12,4))
plt.plot(sunspot['YEAR'], sunspot['SUNACTIVITY'], label='Actual')
plt.plot(pd.Series(range(2009,2030)), forecast, label='Forecast')
plt.xlabel('Year')
plt.ylabel('Sunspot episodes per one year')
plt.legend()
plt.show()
```
