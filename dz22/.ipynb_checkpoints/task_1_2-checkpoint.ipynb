```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn3
math = 15
phys = 16
chem = 19
math_phys = 7
math_chem = 9
phys_chem = 6
everything = 4
only_math_phys = math_phys - everything
only_math_chem = math_chem - everything
only_phys_chem = phys_chem - everything
only_math = math - only_math_chem - only_math_phys - everything
only_phys = phys - only_phys_chem - only_math_phys - everything
only_chem = chem - only_phys_chem - only_math_chem - everything
at_least_one = everything + only_phys_chem + only_math_chem + only_math_phys + only_phys + only_chem + only_math
print(f'Хоча б одну відмінну оцінку має {at_least_one} студентів')
just_one = only_math + only_chem + only_phys
print(f'Тільки одну відмінну оцінку має {just_one} студентів')

venn3(subsets = (only_math, only_phys, only_math_phys, only_chem, only_math_chem, only_phys_chem, everything),
      set_labels = (f'Math - {math}', f'Phys - {phys}', f'Chem - {chem}'))
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn3
rain = 12
wind = 8
cold = 4
rain_wind = 5
rain_cold = 3
wind_cold = 2
everything = 1
only_rain_wind = rain_wind - everything
only_rain_cold = rain_cold - everything
only_wind_cold = wind_cold - everything
only_rain = rain - only_rain_wind - only_rain_cold - everything
only_wind = wind - only_rain_wind - only_wind_cold - everything
only_cold = cold - only_rain_cold - only_wind_cold - everything
total_days = 30
shit_days = only_rain + only_wind + only_cold + rain_wind + rain_cold + wind_cold + everything
good_days = total_days - shit_days
print(f'Гарних днів у вересні було {good_days}')

venn3(subsets = (only_rain, only_wind, only_rain_wind, only_cold, only_rain_cold, only_wind_cold, everything),
      set_labels = (f'Rain - {rain}', f'Wind - {wind}', f'Cold - {cold}'))
plt.show()
```
