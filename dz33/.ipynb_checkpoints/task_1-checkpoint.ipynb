```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


cars = pd.read_excel('cars.xlsx')
cars
```

```python
fig, axs = plt.subplots(1, 3, figsize=(14,5))
sns.lineplot(data=cars, x='Cons', y='Price', ax=axs[0])
axs[0].set_title('Споживання палива від вартості')
sns.scatterplot(data=cars, x='Cons', y='Speed', ax=axs[1])
axs[1].set_title('Споживання палива від максимальної швидкості')
sns.scatterplot(data=cars, x='Cargo', y='Speed', ax=axs[2])
axs[2].set_title('Вантажопідйомність від максимальної швидкості')
```

```python
sns.histplot(x='Speed', data=cars, bins=5)
```

```python
sns.boxplot(x = 'Speed', y = 'Make', data = cars)
```

```python
sns.pairplot(cars)
```
