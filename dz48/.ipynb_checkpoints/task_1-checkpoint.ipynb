```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
from sklearn.preprocessing import PowerTransformer
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('Data_to_Transform.csv')
data
```

```python
fig, ax = plt.subplots(1,4, figsize=(20, 5))
for i in range(len(data.columns)):
    sns.histplot(data[data.columns[i]], ax=ax[i])
```

```python
# логарифмічна трансформація
fig, ax = plt.subplots(1,4, figsize=(17, 5))
for i in range(len(data.columns)):
    sns.histplot(np.log(data[data.columns[i]]), ax=ax[i])
```

```python
# трансформація кубічним коренем
fig, ax = plt.subplots(1,4, figsize=(17, 5))
for i in range(len(data.columns)):
    sns.histplot(np.cbrt(data[data.columns[i]]), ax=ax[i])
```

```python
# трансформація Yeo Johnson'a
fig, ax = plt.subplots(1,4, figsize=(17, 5))
for i in range(len(data.columns)):
    sns.histplot(stats.yeojohnson(data[data.columns[i]]), ax=ax[i])
```

#### Для Moderate Positive Skew і Highly Positive Skew краще підходить трансформація логарифмом.
#### Для Moderate Negative Skew і Highly Negative Skew - трансформаці Yeo Johnson'a.
#### Метод box-cox не примінимий для від'ємних значень.
