```python
# Task 1

a1 = 4 / 7
print(f"Ймовірність, що перший елемент робочий = {round(a1*100,1)} %")

a2 = 3 / 6
print(f"Ймовірність, що другий елемент робочий = {round(a2*100,1)} %")

a3 = 3 / 5
print(f"Ймовірність, що третій елемент робочий = {round(a3*100,1)} %")

a = a1 * a2 * a3
print(f"Ймовірність події, що з трьох елементів один буде не робочий (а інші два робочі) = {round(a*100,1)} %")
```

```python
# Task 2

# Ймовірність спрацювання сигналізаторів
a1 = 0.95
a2 = 0.9

a = a1 + a2 - (a1 * a2)
print(f"Ймовірність того, що з хоча б один сигналізатор спрацює = {round(a*100,1)} %")

# Ймовірність, що обидва не спрацюють
b1 = (1 - a1) * (1 - a2)

# Ймовірність, що обидва спрацюють
b2 = a1 * a2

b = 1 - b1 - b2
print(f"Ймовірність того, що тільки один сигналізатор спрацює = {round(b*100,1)} %")
```
